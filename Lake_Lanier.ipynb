```python
import csv
import ssl
import urllib.request

import pandas as pd

ssl._create_default_https_context = ssl._create_unverified_context
```

The current year can always be found at http://water.sam.usace.army.mil/gage/acf/prob1.txt

Records go back as far as February 8, 1956. Structure of URL for older files is:
http://water.sam.usace.army.mil/gage/acf/prob1- + last 2 digits of year + '.txt'

Data is in .txt format but is separated out in an unusual fashion - varying whitespace rather than purely commas or tabs or newlines. Will try to use pandas to take in the data initially, but will likely just extract the first 28 rows to start. 

```python
current_year = 'http://water.sam.usace.army.mil/gage/acf/prob1.txt'

with urllib.request.urlopen(current_year) as f:
    df_thisyear = pd.read_csv(f, skiprows=[0,1,2,3,36,37,38],
                    delim_whitespace=True,
                    index_col=0,
                    names=['Day', 'October', 'November', 'December', 'January',
                          'February', 'March', 'April', 'May', 'June', 'July',
                          'August', 'September']).fillna('')
```
